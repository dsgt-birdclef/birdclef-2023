```python
import os
import wave
import contextlib
import IPython
```

```python
data_path = "../../data/raw/background_audio"
print("Number of Audio Files:", len(os.listdir(data_path)))
```

```python
size = 0
for track in list(os.listdir(data_path)):
    size += (os.path.getsize(f"{data_path}/{track}"))
print("Avg Track Size: ", size / len(os.listdir(data_path)) / 1000, "KB")
```

```python
durations = []
for track in list(os.listdir(data_path)):
    with contextlib.closing(wave.open(f"{data_path}/{track}", 'r')) as f:
        durations.append(f.getnframes() / float(f.getframerate()))
print("Duration of Tracks: ", set(durations), "seconds")
```

```python
tracks = list(os.listdir(data_path))

IPython.display.display(IPython.display.Audio(f"{data_path}/{tracks[5]}"))
IPython.display.display(IPython.display.Audio(f"{data_path}/{tracks[235]}"))
IPython.display.display(IPython.display.Audio(f"{data_path}/{tracks[1065]}"))

```
